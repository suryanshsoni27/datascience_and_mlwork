```python


assets = ['PG','^GSPC']
data = pd.DataFrame()

for a in assets:
    data[a] = wb.DataReader(a,data_source = 'yahoo', start = '2010-1-1',end = '2016-12-31')['Adj Close']

sec_return = np.log(data/data.shift(1))

cov = sec_return.cov() * 250 
cov 

```

```python
cov_with_market = cov.iloc[0,1]
cov_with_market 
```

```python
market_var = sec_return['^GSPC'].var() * 250
market_var

'''
Beta: 
Bpg = sigmapg,m/sigmam

'''


PG_beta = cov_with_market / market_var
```

```python
PG_beta
```
