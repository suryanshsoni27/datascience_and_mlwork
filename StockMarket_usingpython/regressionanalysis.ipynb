```python
import numpy as np 
import pandas as pd 
from scipy import stats
import statsmodels.api as sm
import matplotlib.pyplot as plt 
```

```python
data = pd.read_excel('Housing.xlsx')
```

```python
data
```

```python
data[['House Price','House Size (sq.ft.)']]
```

```python
X = data['House Price']
y = data['House Size (sq.ft.)']
```

```python
plt.scatter(X,y)
```

```python
X1 = sm.add_constant(X)
reg = sm.OLS(y,X1).fit()
```

```python
reg.summary()
```
